```python
N=int(input("enter an integer"))
z=int(input("enter the size"))
i=N
temp=[]
count=0
#find no. of digits
while(i>0):
    i//=10
    count=count+1
if count==z:
    print(N)
elif z>count:
    for i in range(z-count):
        temp.append("0")
    temp.append(N)
print(''.join(map(str,temp)))
```
